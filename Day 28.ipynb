```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Create a new feature: FamilySize
#Create another feature: IsAlone
#Train Models
             #Train Logistic Regression without Family features (baseline).
             #Train Logistic Regression with FamilySize & IsAlone.
#Compare Results
#Check accuracy, precision, recall, F1-score.
#Which model performs better?
```

```python
data["FamilySize"] = data["SibSp"] + data["Parch"] +1
```

```python
data["IsAlone"] = (data["FamilySize"] ==1).astype(int)
```

```python
new_data = data[["Pclass","Sex","Age","SibSp","Parch","Fare","Embarked","FamilySize","IsAlone","Survived"]].copy()
```

```python
new_data.head(3)
```

```python
new_data.isnull().sum()
```

```python
from sklearn.impute import SimpleImputer
```

```python
si = SimpleImputer(strategy ="mean")
new_data["Age"] = si.fit_transform(new_data[["Age"]])
```

```python
new_data["Embarked"] = new_data["Embarked"].fillna(new_data["Embarked"].mode()[0])   
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
new_data["Sex"] = le.fit_transform(new_data["Sex"])
```

```python
le1 = LabelEncoder()
new_data["Embarked"] = le.fit_transform(new_data["Embarked"])
```

```python
x = new_data.iloc[:,:-1]
y = new_data["Survived"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2,random_state = 42)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

# with Family size

```python
lr = LogisticRegression(max_iter = 1000)
lr.fit(x_train,y_train)
```

```python
y_pred_family = lr.predict(x_test)
```

```python
print("\nðŸ”¹ Enhanced Model Performance (with FamilySize & IsAlone)")
print("Accuracy:", accuracy_score(y_test, y_pred_family))
print("Precision:", precision_score(y_test, y_pred_family))
print("Recall:", recall_score(y_test, y_pred_family))
print("F1 Score:", f1_score(y_test, y_pred_family))
```
