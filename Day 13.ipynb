```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("C:/Users/sumit/Downloads/titanic.csv")
data
```

```python
#Add a column IsAlone: True if FamilySize == 1, else False.

data["FamilySize"] = data["SibSp"] + data["Parch"] + 1
data["IsAlone"] = data["FamilySize"].apply(lambda x: x==1)
data[["IsAlone","FamilySize"]]
```

```python
#Show top 10 passengers who paid the highest fare.

highest_fare = data.sort_values(by = "Fare" ,ascending = False).head(10)
highest_fare[["Name", "Fare", "Pclass", "Sex"]]
```

```python
#List all female survivors under age 30.
survival_female = data[(data["Survived"]==1) & (data["Sex"] == "female") & (data["Age"] < 30 )]
survival_female[["Name","Age","Sex","Survived"]]
```

```python
#Sort the dataset by age in descending order and show top 5 oldest passengers.

oldest_passenger = data.sort_values(by ="Age",ascending = False).head(5)
oldest_passenger[["Name","Age","Survived","Sex"]]
```

```python
#Display all passengers who embarked from ‘C’ and survived
passengers = data[(data["Survived"]==1) & (data["Embarked"] == "C" )]
passengers[["Name","Age","Survived","Sex"]]
```

```python
#Find all rows where passengers are alone and in 1st class.
data["FamilySize"] = data["SibSp"] + data["Parch"] +1
alone_passengers = data[(data["FamilySize"] == 1) & (data["Pclass"] == 1 )]
alone_passengers [["Name","Sex","Age","Survived"]]
```
