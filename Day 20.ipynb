```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("C:/Users/sumit/Downloads/titanic.csv")
data.head(3)
```

```python
new_data = data[["Pclass","Sex","Age","SibSp","Parch","Fare","Embarked","Survived"]].copy()
```

```python
new_data.head(3)
```

```python
new_data.isnull().sum()
```

# Handling missing values

```python
from sklearn.impute import SimpleImputer
```

```python
si = SimpleImputer(strategy="mean")
new_data["Age"] = si.fit_transform(new_data[["Age"]])
```

```python
new_data["Embarked"] = new_data["Embarked"].fillna(new_data["Embarked"].mode()[0])
```

# Encoding 

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
new_data["Embarked"] = le.fit_transform(new_data["Embarked"])
```

```python
le1 = LabelEncoder()
new_data["Sex"] = le.fit_transform(new_data["Sex"])
```

```python
#Train 3 models and evaluate
    #Logistic Regression
    #Random Forest
    #Support Vector Machine
```

```python
x = new_data.iloc[:,:-1]
y = new_data["Survived"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2,random_state = 42)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
```

```python
lr = LogisticRegression(max_iter=2000)
lr.fit(x_train,y_train)
```

```python
x_train.shape , y_train.shape
```

```python
lr.score(x_train,y_train)*100 , lr.score(x_test,y_test)*100
```

```python
rf = RandomForestClassifier(n_estimators =4)
rf.fit(x_train,y_train)
```

```python
rf.score(x_train,y_train)*100 , rf.score(x_test,y_test)*100         #overfitting
```

```python
sv = SVC(kernel = "linear")
sv.fit(x_train,y_train)
```

```python
sv.score(x_train,y_train)*100 , sv.score(x_test,y_test)*100
```

# Cross validation 

```python
from sklearn.model_selection import cross_val_score
```

```python
cross_val_score(LogisticRegression(max_iter = 2000),x,y,cv=5)
```

```python
cross_val_score(RandomForestClassifier(),x,y,cv=5)
```

```python
cross_val_score(SVC(kernel ="linear"),x,y,cv=5)
```
