```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("C:/Users/sumit/Downloads/titanic.csv")
data
```

```python
 #Find the youngest female passenger who survived
female_p = data[(data["Sex"]=="female" ) & (data["Survived"] == 1)]
young_female = female_p.sort_values("Age").head(10)
young_female[["Name","Age","Survived","Pclass"]]
```

```python
#How many male passengers aged between 18 and 40 survived?
male_p = data[(data["Sex"] == "male") & (data["Survived"]==1) & (data["Age"]>=18) &(data["Age"]<= 40)]
print("Total male",len(male_p))
male_p[["Name","Age","Pclass","Sex"]]
```

```python
#Count the number of passengers who had a fare above the average fare.

avg_fare = data["Fare"].mean()
above_avg_fare= data[data["Fare"]>avg_fare]
above_avg_fare[["Name","Sex","Pclass","Fare"]]
```

```python
#Create a bar plot showing the number of survivors by Embarked location

survivor = data[data["Survived"]==1]
survivors_counts = survivor["Embarked"].value_counts().sort_index()

plt.figure(figsize=(5,4))
survivors_counts.plot(kind="bar",color='skyblue', edgecolor='black')
plt.title("Number of survivors by Embarked location")
plt.ylabel("Survivors")
plt.xlabel("Embarked location")
plt.show()
```
